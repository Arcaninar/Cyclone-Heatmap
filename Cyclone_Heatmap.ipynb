# Cyclone Heatmap

```python
import folium
import folium.plugins as plugins
from ipywidgets import interact
import ipywidgets as widgets
```

```python
# Extract cleaned data
data = list()

with open('heatmap_data.txt', 'r') as f:
    for line in f:
        numLine = int(line)
        temp = list()
        for i in range(numLine):
            newLine = [float(x) for x in f.readline().replace('\n', '').split(',')]
            temp.append(newLine)
        data.append(temp)
```

```python
# Create year and season list
season_index = list()
season = ['Spring', 'Summer', 'Autumn', 'Winter']

for i in range(2000, 2023):
    for j in season:
        season_index.append(str(i) + ' - ' + j)
```

```python
# Timelapse of the heatmap
map_obj = folium.Map(location = [36.2048, 138.2529], zoom_start = 5)

hmwt = plugins.HeatMapWithTime(
    data,
    index = season_index,
    auto_play = True,
    radius = 25,
    blur = 0.8,
    max_opacity = 1,
    min_opacity = 0.1,
    position = 'topright',
    gradient = {0 : 'navy', 0.6 : 'cyan', 0.7 : 'lime', 0.8 : 'yellow', 1 : 'red'}
)

hmwt.add_to(map_obj)

folium.raster_layers.TileLayer('Stamen Terrain').add_to(map_obj)

map_obj
```

```python
# Heatmap selector for each year and season
@interact(
    season = widgets.SelectionSlider(description='Season',   options=['Spring', 'Summer', 'Autumn', 'Winter']),
    year = (2000, 2022)
)

def test(season, year=2000):
    seasonList = ['Spring', 'Summer', 'Autumn', 'Winter']
    map_slider = folium.Map(location = [36.2048, 138.2529], zoom_start = 5)
    
    plugins.HeatMap(
        data[seasonList.index(season) + 4*(year - 2000)],
        radius = 25,
        blur = 15,
        gradient = {0 : 'navy', 0.6 : 'cyan', 0.7 : 'lime', 0.8 : 'yellow', 1 : 'red'}
    ).add_to(map_slider)
    
    folium.TileLayer('stamenterrain').add_to(map_slider)
    
    display(map_slider)
```

## Cyclone Analysis

We can see that most of the Cyclone happened during Summer & Autumn, and those cyclones always come from either the South of Japan or East of Japan.
