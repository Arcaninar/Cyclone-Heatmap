The period from March to May is spring, June to August is summer, September to November is autumn, and December to February is winter

```python
import folium
import folium.plugins as plugins
from ipywidgets import interact
import ipywidgets as widgets
```

```python
data = list()

with open('heatmap_data.txt', 'r') as f:
    for line in f:
        numLine = int(line)
        temp = list()
        for i in range(numLine):
            newLine = [float(x) for x in f.readline().replace('\n', '').split(',')]
            temp.append(newLine)
        data.append(temp)
```

```python
season_index = list()
season = ['Spring', 'Summer', 'Autumn', 'Winter']

for i in range(2000, 2023):
    for j in season:
        season_index.append(str(i) + ' - ' + j)
```

```python
map_obj = folium.Map(location = [36.2048, 138.2529], zoom_start = 5)

hmwt = plugins.HeatMapWithTime(
    data,
    index = season_index,
    auto_play = True,
    radius = 25,
    blur = 0.8,
    max_opacity = 1,
    min_opacity = 0.1,
    position = 'topright',
    gradient = {0 : 'navy', 0.6 : 'cyan', 0.7 : 'lime', 0.8 : 'yellow', 1 : 'red'}
)

hmwt.add_to(map_obj)

folium.TileLayer('stamenterrain').add_to(map_obj)

map_obj
```

```python
@interact(
    season = widgets.SelectionSlider(description='Season',   options=['Spring', 'Summer', 'Autumn', 'Winter']),
    year = (2000, 2022)
)

def test(season, year=2000):
    seasonList = ['Spring', 'Summer', 'Autumn', 'Winter']
    map_slider = folium.Map(location = [36.2048, 138.2529], zoom_start = 5)
    
    plugins.HeatMap(
        data[seasonList.index(season) + 4*(year - 2000)],
        radius = 25,
        blur = 15,
        gradient = {0 : 'navy', 0.6 : 'cyan', 0.7 : 'lime', 0.8 : 'yellow', 1 : 'red'}
    ).add_to(map_slider)
    
    folium.TileLayer('stamenterrain').add_to(map_slider)
    
    display(map_slider)
```

```python
seasonList = ['Spring', 'Summer', 'Autumn', 'Winter']
for i in range(2000, 2023):
    for j in range(4):
        m_i = folium.Map(location = [36.2048, 138.2529], zoom_start = 5)

        # Add year label to the map
        title_html = '''
                     <h3 align="center" style="font-size:22px"><b>{}</b></h3>
                     '''.format('Season: ' + str(i) + ' - ' + seasonList[j])   
        m_i.get_root().html.add_child(folium.Element(title_html))

        # Plot colors to the map using json geographical borders, and total courses from total_perYear dict
        plugins.HeatMap(
            data[j + 4*(i - 2000)],
            radius = 25,
            blur = 15,
            gradient = {0 : 'navy', 0.6 : 'cyan', 0.7 : 'lime', 0.8 : 'yellow', 1 : 'red'}
        ).add_to(m_i)

        folium.TileLayer('stamenterrain').add_to(m_i)


        m_i.save(str(i) + '_' + str(j) + '.html')
```
